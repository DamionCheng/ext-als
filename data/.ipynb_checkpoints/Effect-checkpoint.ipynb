下面是loss变化过程的绘图，由于两种方法的loss曲线完全重合，故在此分别绘制。

````python
import re
def getLoss(fname):
    loss = []
    with open(fname) as f:
        for l in f.readlines():
            if l.startswith('Iter'):
                v = re.findall('loss: (\d+.\d+)', l)
                loss.append(float(v[0]))
    return loss
````

````python
ext_f1 = getLoss('extALS.log')
ext_f5 = getLoss('extALS_f5.log')
e_f1 = getLoss('eALS.log')
e_f5 = getLoss('eALS_f5.log')
````

````python
import matplotlib.pyplot as plt

#fig, axs = plt.subplots(1,2)

xais = range(500)
plt.plot(xais, ext_f1, 'r')
plt.show()
plt.plot(xais, e_f1)
plt.show()
plt.plot(xais, ext_f5, 'r')
plt.show()
plt.plot(xais, e_f5)
plt.show()
````
